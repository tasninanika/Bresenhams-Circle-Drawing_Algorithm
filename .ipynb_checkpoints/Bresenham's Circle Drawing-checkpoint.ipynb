```python
import matplotlib.pyplot as plt

def bresenham_circle(cx, cy, r):
    x, y = 0, r
    p = 3 - 2 * r
    points = []

    # Function to plot symmetrical points
    def plot_circle_points(cx, cy, x, y):
        points.extend([(cx + x, cy + y), (cx - x, cy + y), (cx + x, cy - y), (cx - x, cy - y),
                       (cx + y, cy + x), (cx - y, cy + x), (cx + y, cy - x), (cx - y, cy - x)])

    # Plot the initial points
    plot_circle_points(cx, cy, x, y)

    while x < y:
        x += 1
        if p <= 0:
            p = p + 4 * x + 6
        else:
            y -= 1
            p = p + 4 * (x - y) + 10
        plot_circle_points(cx, cy, x, y)

    return points

def plot_graph(circle_points):
    # Plot the circle points on a graph
    if circle_points:
        x_vals, y_vals = zip(*circle_points)
        plt.scatter(x_vals, y_vals, color='r', s=40)  # Plot each point as red
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Bresenham's Circle Algorithm")
    plt.grid(True)
    plt.axis("equal")  # Equal scaling of X and Y axes
    plt.show()

# Example usage
cx, cy, r = 5, 5, 4
circle_points = bresenham_circle(cx, cy, r)
plot_graph(circle_points)
```
